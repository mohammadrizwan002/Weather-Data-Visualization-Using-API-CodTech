```python
# Task 1 - CodTech Internship
# Weather Forecast Visualization using OpenWeatherMap API

import requests
import matplotlib.pyplot as plt

# --------------- CONFIGURATION ---------------
API_KEY = '3af61c5e3aa0d5c96ad8c2d99ae1fd17'  #  Your latest working API key
city = 'Hyderabad'
units = 'metric'
url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_KEY}&units={units}"

# --------------- FETCH DATA FROM API ---------------
response = requests.get(url)
data = response.json()

# --------------- HANDLE ERRORS IF ANY ---------------
if 'list' not in data:
    print(" ERROR: Unable to fetch data. Check API key or city name.")
    print(" API Response:", data)
else:
    # --------------- EXTRACT DATE AND TEMPERATURE ---------------
    dates = []
    temperatures = []

    for item in data['list']:
        temp = item['main']['temp']
        date = item['dt_txt']
        temperatures.append(temp)
        dates.append(date)

    # --------------- VISUALIZE USING MATPLOTLIB ---------------
    plt.figure(figsize=(12, 5))
    plt.plot(dates[:8], temperatures[:8], marker='o', color='blue', label='Temperature')
    plt.xticks(rotation=45)
    plt.xlabel("Date & Time")
    plt.ylabel("Temperature (C)")
    plt.title(f"Weather Forecast for {city}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # --------------- SAVE OUTPUT CHART ---------------
    plt.savefig("output_chart.png")  #  Saves chart as PNG
    plt.show()
```
